```python
columns_mix = ['Confirmed','Deaths','Recovered','Active']
# Calculate the sum of specified columns
columns_add = df[columns_mix].sum().plot(kind='bar')
```

```python
region=df.groupby(by='WHO Region').sum()
region
```

```python
plt.figure(figsize=(10,5))
region['Confirmed'].plot(kind="bar")
plt.title('Confirmed_Cases')
```

```python
plt.figure(figsize=(10,6))
region['Deaths'].plot(kind="line")
plt.title('Deaths_cases')
```

```python
region['Recovered'].plot(kind="pie")
plt.title('Recovered_cases')
```

```python
region['Active'].plot(kind="pie")
plt.title('Active_cases')
```
